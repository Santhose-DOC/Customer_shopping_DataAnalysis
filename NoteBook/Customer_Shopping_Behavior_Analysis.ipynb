```python
import pandas as pd
df=pd.read_csv("customer_shopping_behavior.csv")
```

```python
df.describe(include="all")
```

```python
#we will find out median within each category and replacedm for each acategorical null values

df["Review Rating"]=df.groupby("Category")['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
df.columns=df.columns.str.lower()
df.columns=df.columns.str.replace(' ','_')
```

```python
df.columns
```

```python
df=df.rename(columns={"purchase_amount_(usd)":"purchase_amount"})
```

```python
df.columns
```

```python
lables=["Young Adults","Adult","middle-aged","Senior"]
df["age_group"]=pd.qcut(df["age"],q=4,labels=lables)
```

```python
df[["age_group","age"]].head(8)
```

```python
a=df["frequency_of_purchases"].head(15)
a
```

```python
# Create column purchase_frequency_days to Converts the frequencey of datas as no.of days
frequency_mappings={
    "Fortnightly":14,
    "Weekly":7,
    "Monthly":30,
    "Bi-Weekly":14,
    "Quarterly":90,
    "Annually":365,
    "Every 3 Months":90
}
df["purchase_frequency_days"]=df["frequency_of_purchases"].map(frequency_mappings)
```

```python
df[["purchase_frequency_days","frequency_of_purchases"]].head(10)
```
